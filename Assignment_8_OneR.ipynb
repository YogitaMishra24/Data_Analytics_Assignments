```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]


train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
def one_r(df, target):
    attributes = [col for col in df.columns if col != target]
    best_attr, best_rule, min_error = None, {}, float('inf')

    for attr in attributes:
        rule = {}
        total_error = 0
        df[attr] = pd.cut(df[attr], bins=3, labels=['low', 'medium', 'high'])

        for value in df[attr].unique():
            subset = df[df[attr] == value]
            most_common_class = subset[target].mode()[0]
            rule[value] = most_common_class
            total_error += sum(subset[target] != most_common_class)

        if total_error < min_error:
            min_error = total_error
            best_attr = attr
            best_rule = rule

    return best_attr, best_rule, min_error


best_attr, best_rule, min_error = one_r(train_df.copy(), 'species')


print(" Best Attribute:", best_attr)
print(" Generated Rule:")
for k, v in best_rule.items():
    print(f"  If {best_attr} = {k} → {v}")
print("Total Classification Errors:", min_error)
```
