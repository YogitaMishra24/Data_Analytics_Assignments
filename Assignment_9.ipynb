```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Create a dataset with 2D points (x, y)
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Sample (x, y) Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
wcss = []  # within-cluster sum of squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS (Inertia)")
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 1: Create a dataset (x, y)
X, y = make_blobs(
    n_samples=300,      # total data points
    centers=4,          # true clusters
    cluster_std=0.60,   # standard deviation of clusters
    random_state=0
)

# Visualize dataset
plt.scatter(X[:, 0], X[:, 1], s=30, c='gray')
plt.title("Generated Dataset (x, y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Step 2: Apply Elbow Method to find optimal K
inertia = []  # list to store SSE (sum of squared errors)
K_values = range(1, 11)

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # store SSE value

# Step 3: Plot Elbow curve
plt.plot(K_values, inertia, 'bo-')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia / SSE")
plt.show()

# Step 4: Choose k (from the elbow point, usually around 4)
k_optimal = 4
kmeans = KMeans(n_clusters=k_optimal, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Step 5: Visualize final clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', marker='X', label='Centroids')
plt.title(f"K-Means Clustering (k={k_optimal})")
plt.legend()
plt.show()
```
