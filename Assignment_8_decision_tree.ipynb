```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the Iris flower dataset
iris = load_iris()

# Feature (X) and target (y)
X = iris.data
y = iris.target

print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)

# Create a Decision Tree model
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)

# Train (fit) the model
model.fit(X, y)

# Predict for the first 5 samples
predictions = model.predict(X[:5])
print("Predictions:", predictions)
print("Actual:", y[:5])

plt.figure(figsize=(12, 8))
plot_tree(model, 
          filled=True, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names,
          rounded=True)
plt.show()

# Example: predict flower type for new measurements
sample = [[5.1, 3.5, 1.4, 0.2]]  # Sepal & petal length/width
predicted_class = model.predict(sample)
print("Predicted class:", iris.target_names[predicted_class][0])
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# ID3 method
id3_model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
id3_model.fit(X, y)

# Visualization
plt.figure(figsize=(10, 6))
plot_tree(id3_model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree using ID3 (Entropy)")
plt.show()
```

```python
cart_model = DecisionTreeClassifier(criterion='gini', max_depth=3)
cart_model.fit(X, y)

plt.figure(figsize=(10, 6))
plot_tree(cart_model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree using CART (Gini Index)")
plt.show()
```
